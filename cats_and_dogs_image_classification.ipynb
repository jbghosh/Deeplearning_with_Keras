# packages and access to the image folder

```python
import numpy as np
import cv2
#pip install opencv-python
import os #spacify the location per images
import random
import matplotlib.pyplot as plt
import pickle
%matplotlib inline
```

# Make the images to an uniform 2D array

```python
img_size = 100

data = []

for catagory in categories:
    folder = os.path.join(Directory, catagory)
    label= categories.index(catagory)
    for img in os.listdir(folder):
        try:
            img_path= os.path.join(folder, img)
            img_arr= cv2.imread(img_path, cv2.IMREAD_GRAYSCALE) # make the image black and white
            img_arr= cv2.resize(img_arr, (img_size, img_size))
            data.append([img_arr, label])

        except Exception as e:  # in the interest in keeping the output clean...
                pass           
```

```python
len(data)
```

```python
random.shuffle(data)
```

```python
data[0]
```

```python
x = []
y=[]

for features, labels in data:
    x.append(features)
    y.append(labels)
```

```python
x= np.array(x)
y=np.array(y)
```

```python
pickle.dump(x, open('x.pkl', 'wb')) #wb is write in binary
pickle.dump(y, open('y.pkl', 'wb'))
```

# Training

```python
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
```

```python
model = Sequential()

model.add(Conv2D(64, (3,3),  activation = 'relu')) #64 feature detecters , each filetrs are of 3x3 dimension  
model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Conv2D(64, (3,3), activation = 'relu'))
model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Flatten())

model.add(Dense(128, input_shape= x.shape[1:], activation='relu'))

model.add(Dense(2, activation= 'softmax'))
```

```python
#loss functions and optimizers
```

```python
model.compile(optimizer='adam', loss= 'sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit(x,y, epochs=5, validation_split= 0.1)
```

```python
model.save('cats_and_dogs.model')
```
